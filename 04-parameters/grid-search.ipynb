# Grid Search - Random Forest

Experiments using Random Forest that provides the best performance in the Weka experiment.

## Load Datasets

```python
import pandas as pd
```

```python
# normal dataset
dataset = pd.read_csv('../03-features-labels/dataset.csv')
dataset.head()
```

```python
# scaled dataset
dataset_scaled = pd.read_csv('../03-features-labels/dataset_scaled.csv')
dataset_scaled.head()
```

### Specify Features

```python
feature_columns = "RectCoverage,AvgRectArea,RectStDev,AspectRatio,AvgLineBends,AvgLineLength,LongestLine,ShortestLine,LineLengthStDev,AvgLineAngle,OrthLinesRatio,LineCrossings,AvgCrossingAngle,AvgShortestDistance,RectDistribution,RectOrth2,rectangles,lines".split(',')
```

```python
# normal dataset
dropcolumns = (c for c in dataset.columns if c not in feature_columns)
features = dataset.drop(columns=dropcolumns)
features.info()
```

```python
# scaled dataset
dropcolumns = (c for c in dataset_scaled.columns if c not in feature_columns)
features_scaled = dataset_scaled.drop(columns=dropcolumns)
features_scaled.info()
```

### Specify Labels

```python
label_column = "Quality"
# labels are the same for both dataset since only features were scaled
labels = dataset[label_column]
labels.head()
```

## Experiments

### Define Grid Search Parameters

```python
from sklearn.ensemble import RandomForestRegressor
import sklearn.metrics as m
from sklearn.model_selection import KFold 
from sklearn.model_selection import StratifiedKFold

# Defining variables for the experiments

RANDOM_SEED = 41

classes = [f"{(x+2)/2:.1f}" for x in range(9)]
y_cat = pd.cut(labels, [0,1,1.5,2,2.5,3,3.5,4,4.5,5], labels=classes)
strat = StratifiedKFold(10)
folding = KFold(n_splits=10, shuffle=True, random_state=RANDOM_SEED)

rf_parameters = {
  'regressor__max_depth':[None,4,5,6,8,10],
  'regressor__criterion':['squared_error', 'absolute_error', 'poisson'],
  'regressor__n_estimators':[10, 100, 500],
  'regressor__max_features':['auto', 'sqrt', 'log2']
}

regressors = {'RandomForest': RandomForestRegressor(random_state=RANDOM_SEED)}

regressors_parameters = {'RandomForest': rf_parameters}

scoring = {
  'mae' : m.make_scorer(m.mean_absolute_error),
  'mse' : m.make_scorer(m.mean_squared_error, squared=True),
  'rmse': m.make_scorer(m.mean_squared_error,squared=False),
  'r2'  : m.make_scorer(m.r2_score)
}

features = {
  'regular': features,
  'scaled': features_scaled
}
```

### Run the experiments

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from datetime import datetime

results = pd.DataFrame()

start = datetime.now()

for feature in features.keys():
  for regressor in regressors.keys():
    print("## Experiments using {} with {} features ##".format(regressor, feature))
    
    steps = []
    steps.append(('regressor', regressors[regressor]))
    pipeline = Pipeline(steps)
    
    skf_gen = strat.split(features[feature], y_cat)
    experiments = GridSearchCV(pipeline, regressors_parameters[regressor], cv=skf_gen, scoring=scoring, refit=False)
    experiments.fit(features[feature], labels)
    
    exp_params = "{}-{}".format(regressor, feature)
    results = results.append(pd.DataFrame().from_dict(experiments.cv_results_).assign(Method=exp_params))
        
# save to csv
results.to_csv('grid-search-result.csv', sep=',')

finish = datetime.now()
time_needed = finish - start

print("## Finished in {} ##".format(time_needed))
```
